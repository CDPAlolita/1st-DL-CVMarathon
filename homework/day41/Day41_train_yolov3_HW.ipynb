A: 資料集的格式分為圖檔(image)、種類名稱檔(class names file)和標註檔(annotation file)
圖檔: 就是一般的圖片。
標註檔: 每一列開頭為圖片的路徑 接著為該圖中各物件bounding box的座標及總類，
       各bounding box之間以空格分開，
       bounding box內的座標、種類資訊則由","分隔，不可用空格。
種類名稱檔: 全檔只有一列，列出所有種類的名稱，各類間以空格分隔。
